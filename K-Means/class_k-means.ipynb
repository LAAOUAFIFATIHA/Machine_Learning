# K-means
K-means est un **algorithme de clustering (classification non supervisée)** qui permet de regrouper des données en k clusters (groupes) sur la base de leur similarité. Il est souvent utilisé en analyse de données, segmentation de clients, compression d'images, etc.

### Principe de K-means

    1- Choix du nombre de clusters k : L’utilisateur définit k, le nombre de groupes souhaités.
    2- Initialisation : On choisit k centres de clusters initiaux.
    3- Assignation des points : Chaque point est affecté au centre le plus proche (en utilisant par exemple la distance euclidienne).
    4- Mise à jour des centres : On recalcule les centres des clusters comme la moyenne des points assignés.
    5- Répétition : On répète les étapes 3 et 4 jusqu'à convergence (c'est-à-dire jusqu'à ce que les centres ne changent plus ou qu'un critère d'arrêt soit atteint).
    

# K-MEANS

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
import matplotlib.image as mpimg
```

```python
class k_means:
    def __init__(self,):
        pass

    # destance    
    def dest(self ,S , c ):
        return np.sqrt( np.sum(  (S - c) **2 , axis =1 ) )
        
    # train
    def fit(self, X , k , itera):
        n,m=X.shape
        centroids=X[np.random.choice(n, k)]
        mat=np.zeros((n,k+1)) 
        for _ in range(itera):
            for i in range(k):
                mat[:,i] = self.dest(X , centroids[i])
            mat[:,k] = np.argmin(mat[:,:k] , axis=1)
            for j in range(k):
                # fliter mat[ : , k ] == j    
                centroids[j] = np.mean( X[mat[  : , k ] == j ] , axis=0 )
        return centroids , mat 
        
    def show_graph(self, X,mat,k , centro ):
        for i in np.unique(mat[:,k]):
            c = X[ mat[:,k]==i] 
            plt.scatter(c[:,0],c[:,1])
            for i in range(centroides.shape[0]):
                plt.scatter(centroides[i][0], centroides[i][1])
                
    def fit_img(self, img_path ,k ):
        img=mpimg.imread(img_path)
        # Show the original image before clustering
        plt.figure(figsize=(10, 5))
        plt.subplot(1, 2, 1)
        plt.imshow(img)
        img.shape
        a,b,c=img.shape
        M=img.reshape((-1,3))/255
        centroids ,mat= self.fit(M,k,10)
        for i in range(k):
          M[mat[:,k]==i]=centroids[i]
        MM=M.reshape((a,b,c))
        plt.subplot(1, 2, 2)
        plt.imshow(MM) 
        

    def calculate_sse(self ,X, centroids):
        sse = 0
        for i in range(X.shape[0]):  
            distances = np.linalg.norm( X[i] - centroids, axis=1) ** 2
            sse += np.min(distances)
        return sse
```

```python
KM = k_means() 
```

```python
X, y = make_blobs(n_samples=100, centers=3, n_features=2,random_state=0)
#normalement y est inconue (il ne sera pas utilisé)
# plt.scatter(X[:,0],X[:,1])
```
