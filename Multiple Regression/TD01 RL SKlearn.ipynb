##  Exercice 1 :  Prédiction du Prix des Maisons

Vous êtes un data scientist travaillant pour une agence immobilière. Votre objectif est de
prédire le prix des maisons en fonction de leur superficie en m².


1. Importer les bibliothèques nécessaires.

```python
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

2. Créer et visualiser les données.

```python
x = np.array([30, 50, 70, 90, 110, 150])
y = np.array([75, 100, 130, 180, 210, 250])
plt.scatter(x , y )
```

3. A l’aide du modèle LinearRegression de sklearn.linear_model, effectuer une régression linéaire pour modéliser la relation entre superficie et prix.


```python
model = LinearRegression()
# because fit need the dim of x to be 2 
x_reshaped = x.reshape(-1 , 1 )
model.fit(x_reshaped,y)
```

4. Afficher l'équation de la droite trouvée (y = a*x + b).

```python
w = np.array([  model.intercept_ ,  model.coef_[0] ])
y_pred = w[1]*x + w[0]
plt.scatter(x , y )
plt.plot(x,y_pred , color='r')
```

5. Faire une prédiction du prix pour une maison de 120m²

```python
model.predict([[120]])
```

6. Évaluer la performance du modèle avec le coefficient de détermination R² (utiliser la fonction r2_score() de sklearn.metrics ).

```python
r2_score(y, y_pred)
```

Vous êtes un data scientist travaillant pour une agence immobilière. Votre objectif est de
prédire le prix des maisons en fonction de leur superficie en utilisant une régression
polynomiale.

1. Importer les bibliothèques nécessaires.


```python
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
```

2. Visualiser les données sous forme de nuage de points

```python
x = np.array([30, 50, 70, 90, 110, 150])
y = np.array([75, 100, 130, 180, 210, 300])
plt.scatter(x , y )
```

3. A l’aide de PolynomialFeatures de sklearn.preprocessing et LinearRegression Appliquer une régression polynomiale (de degré 2 ou 3).


```python
x = np.array([30, 50, 70, 90, 110, 150])
y = np.array([75, 100, 130, 180, 210, 300])
plt.scatter(x , y )

d = 2
x_reshaped = x.reshape(-1 , 1)
poly = PolynomialFeatures(degree=d)
x_poly = poly.fit_transform(x_reshaped)

print(x_reshaped)

model_P = LinearRegression()
model_P.fit(x_poly , y)
print(model_P.coef_)
model_P.intercept_
```

```python
model_P.predict(x_poly)
```

4. Évaluer la performance du modèle avec le coefficient de détermination R²


```python
w = np.array( [ model_P.coef_[0]  ,model_P.coef_[1]  ,model_P.coef_[2]  ]) 
y_pred = model_P.intercept_ + x_poly.dot(w)
print(y_pred)
r2_score( y_pred ,y)
```

5. Prédire le prix d’une maison de 120 m² avec le modèle polynomial.

```python
model_P.predict([[ 1,120,120*120]])
```
