# Moidre carre
c'est une methode direct pour oubtenir les meilleures W

```python
# la fonction predictive qui return une probabilit√©
def h(w ,x):
    return w[0]+w[1]*x

# la focntion cout
def cout(w,x,y):
    return ((h(w,x)-y).dot((h(w,x)-y))) / x.shape[0]*2

# moyenne 
def moyenne(x):
    return sum(x)/x.shape[0]

# variance
def variance(x):
    return x.dot(x)/x.shape[0] - np.power(x.mean(),2)

# l'ecartype
def ecart(x):
    return np.sqrt(variance(x))

# covariance
def cov(x , y):
    return (x.dot(y)/x.shape[0]) - (x.mean() * y.mean())

def corr(x , y):
    return cov(x,y)/(ecart(x)*ecart(y))

def moidre_carre(x , y):
    w[1] =  cov(x,y)/(np.power(ecart(x),2)) 
    w[0] =  moyenne(y) - w[1]*moyenne(x)
    return w
# w = [1,8]  
```

```python
w = [1,2]
x = np.array([2,4,14,8,12])
y = np.array([10 ,8 ,2,6,4])
w = moidre_carre(x,y)
y1 = w[0]+w[1]*x
plt.plot(x,y1)
plt.scatter(x,y)
print("l'erreur: ",cout(w,x,y))
print("la correlation entre x et y est =",corr(x,y))
w
```

# Descente de gradient

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Fonction pr√©dictive
def h(w, x):
    return w[0] + w[1] * x

# Fonction de co√ªt
def cout(w, x, y):
    return np.sum((h(w, x) - y) ** 2) / (2 * x.shape[0])

# Descente de gradient
def GradDecente(x, y, alpha, num_iterations=600):
    J = []
    w = [1, 2]  
    for i in range(num_iterations):
        w[1] = w[1] - alpha * (1 / x.shape[0]) * np.sum((h(w, x) - y) * x)
        w[0] = w[0] - alpha * (1 / x.shape[0]) * np.sum(h(w, x) - y)
        J.append(cout(w, x, y))
    return w, J
```

```python
# Donn√©es
x = np.array([1, 2, 3, 4 , 5 , 6 ,7])
y = np.array([8, 7, 6, 4 ,3 ,2 ,1])

# Param√®tres
alpha = 0.017

# Descente de gradient
w, J = GradDecente(x, y, alpha)

# Pr√©diction avec les poids finaux
y1 = w[0] + w[1] * x

# Tracer les points de donn√©es et la droite de r√©gression
plt.scatter(x, y, color='blue', label='Donn√©es originales')
plt.plot(x, y1, color='red', label='R√©gression lin√©aire') 
plt.xlabel('x')
plt.ylabel('y')
plt.title('R√©gression lin√©aire par descente de gradient')
plt.legend()
plt.grid(True)
plt.show()


print("LES COEF W:", w)
print("Co√ªt final:", J[-1])
```

# Autre M√©thode 

## $$ w = \left( X^T X \right)^{-1} X^T Y $$

```python
from scipy import linalg
```

```python
x = np.array([2,4,14,8,12])
y = np.array([10 ,8 ,2,6,4])

ones = np.ones(5) 
X = np.stack((ones,x), axis=1) 
X.T
```

```python
w = linalg.inv(X.T.dot(X)).dot(X.T).dot(y)  #<------ tres important
```

```python
Y = w[0] + w[1]*X
plt.scatter(x,y)
plt.plot(X,Y)
plt.title("Nouvelle M√©thode")
plt.grid(True)
plt.legend()
```

# La Regression lineare simple Avec SKLEAN

```python
import numpy as np
from numpy import random
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt 
```

```python
x  = np.linspace(1,10 , 100)
y  = 3*x +2 + random.rand(x.shape[0])*10-5

model = LinearRegression()
x_train , x_test , y_train , y_test = train_test_split( x , y , test_size = 0.33 , random_state=42 )

# the fit methode take the x input as a 2D array  {1D ----> 2D}
x_train = x_train.reshape(-1, 1)  
x_test = x_test.reshape(-1, 1)

# trainging the model by fit 
model.fit(x_train , y_train)

# preparing the x to the plot it should be as the same size as the value returned by predict 
x1 = np.array(x_test[:,0])
plt.scatter(x,y , color='b')
plt.plot(x1,model.predict(x_test) , color="r")
# x_test
```

```python
s = model.score(x_test , y_test)
print(s)
```
